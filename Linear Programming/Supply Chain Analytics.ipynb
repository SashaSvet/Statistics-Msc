# Supply Chain Optimization 

Supply chain optimization involves finding the best path to achieve an objective based on particulr constraints. 
One of the most powerful tools for Optimization is Linear Programming. 

Basic Components in LP : 

1. Decision Variables - What we (can) control 
2. Objective function - What we want to minimize/maximize - our goal of optimization 
3. Constraints - set of equations that describe the limits of our solution. 



1. Linear Programing (LP)         -  Decision Variables : Only Continuous.
2. Integer Programing (IP)        -  Decision Variables :Only Discrete or Integers.
3. Mixed Integer Programing (MIP) -  Decision Variables :Mix of Continuous and Discrete.
