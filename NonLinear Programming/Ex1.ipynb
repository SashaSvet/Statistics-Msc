## i

    Use the bisection method to find a positive root of the equation
    x = 2sinx:
    accurate to six signifcant digits.

```python
""" Function for x """
def f(x):
    return  x- 2*np.sin(x)
```

```python
""" Bisection Method """


def BisectionMethod(a,b): 
    # Input Check
    if (f(a) * f(b) >= 0): 
        print("Please input compatible a and b") 
        return
    count = 0 
    c = a 
    while ((b-a) >= 0.000001): 
        c = (a+b)/2
        count = count +1 
        if (f(c) == 0.0): 
            break
        # Continue
        if (f(c)*f(a) < 0): 
            b = c 
        else: 
            a = c               
    print("The value of root is : ","%.6f"%c) 
    
    return print('Number of Iterations:',count)
      
a = 1.57
b = 2
BisectionMethod(a, b)
```

## (ii)

Use the Newton method to find a positive root of the equation
x = 2sinx:

```python
def DFunc( x ): 
    return 1 - 2 * np.cos(x)
  
# Function to find the root 
def newton( x ): 
    ratio = f(x) / DFunc(x)
    count = 0
    while abs(ratio) >= 0.000001: 
        ratio = f(x)/DFunc(x) 
        
        x = x - ratio
        count = count + 1
      
    print("The value of the root is : ", 
                             "%.6f"% x) 
    return print('Number of Iterations:',count)
  

x0 = 1.57 
newton(x0)
```

```python
def f(x):
    return (-(1/x)-(2**(0.5)))

def DFunc(x):
    return 1/x**2
```

```python
# דרך טיפה שונה מהדרך הראשונה של הקוד של שיטת ניוטון, שאפשר להגדיר מספיר אטרציות מקסימלי

def ModifiedNewton(x0,iterations):  
    
    # Input check
    if x0 >= 0 :
        return print('Error, please enter the correct value')    
    xn = x0
    for n in range(0,iterations):
        fxn = f(xn)
        if abs(fxn) < 0.00001:
            print('found solution')
            return abs(xn)
        # נוסחת נסיגה
        xn = 2*(x0) + (np.sqrt(2))*((x0)**2)
        if abs(xn - x0) >= 0.00001:
            x0 = xn
        else:
            x0 = x0 
        print(n,abs(xn))
        
    print('Exceeded maximum iterations. No solution found.')
    return None

x0 = -1
ModifiedNewton(x0,1000)
```
